# Purpose of this Program
All the videos within the UMiami dataset are in **wmv** format, which is not viewable in some devices like MacOS. This program converts all wmv videos into **mp4** format.

```python
# Connect to Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import Necessary Libraries
import glob
from moviepy.editor import VideoFileClip

# Convert and save a video from one format to another format
# To convert to mp4, make sure output_file ends with .mp4
def convert_to_mp4(input_file, output_file):
    video = VideoFileClip(input_file)
    video.write_videofile(output_file,codec='libx264', audio_codec='aac')

current_directory = "/content/drive/MyDrive/Deception_Detection/Datasets/MU3D-Package"
source = current_directory + "/Videos"
destination = current_directory + "/VideosMP4"

for file in glob.glob(source + "/*"):
    filename = file.split("/")[-1][:-4]
    if file[-4:] == ".wmv":
        output_file = destination + "/" + filename + ".mp4"
        convert_to_mp4(file, output_file)
```
